```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
output = [3, 0, 4, 0, 5, 0, 6, 0, 7, 11, 8, 330, 9, 3242, 10, 4235, 11, 4240, 12, 4243, 13, 4244, 14, 4244, 15, 4244, 16, 4244, 17, 4244, 18, 4244, 19, 4244, 20, 4244, 50, 4241, 100, 4233, 500, 3212, 1000, 1605, 10000, 0]
dnp = np.reshape(output, (-1, 2)).T
enp = []
# number of genes in e.coli:
for number in dnp[1]:
    enp.append(number * 100 / 4289)
dnp[1] = np.array(enp)
df = pd.DataFrame({"k": dnp[0], "unique_genes_%" : dnp[1]})
#print(df)

sns.set(font_scale = 3.5)
plt.figure(figsize = (50, 20))

ax = sns.barplot(x = "k", y = "unique_genes_%", data = df, color = "Lightblue")
ax.set_title("Unique Kmer Gene Identification in e.Coli\n")
sns.set_context("talk")
ax.get_figure().savefig("eColi_plot_a.png")
```
